```python
import numpy as np
import scipy
import os

import librosa
from librosa import cqt
from librosa.core import amplitude_to_db
import librosa.display

import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
waveform, sr = librosa.load(
    'octave_glissando.wav', sr=None)
hop_length = 512

scalogram= librosa.core.cqt(y=waveform, 
                            bins_per_octave=24, 
                            hop_length=hop_length, 
                            n_bins=24*8, 
                            sr=sr, 
                            sparsity=1e-7)
scalogram= np.abs(scalogram)**2

plt.figure(figsize=(2.5,2.5))
librosa.display.specshow(data=amplitude_to_db(scalogram), 
                         cmap='magma_r',
                         sr=sr,
                         hop_length=hop_length,
                         x_axis='time',
                         bins_per_octave=24,
                         y_axis='cqt_note')
plt.ylim(200, 5000)
plt.xticks([0, 1, 2, 3, 4])
plt.xlabel("Time (s)")
plt.yticks(262 * np.array([1, 2, 4, 8, 16]))
plt.minorticks_off()
plt.ylabel("Log-frequency")

if os.path.isfile('octave_glissando.pdf'):
    os.remove('octave_glissando.pdf') 
plt.savefig('octave_glissando.pdf', bbox_inches='tight')
```
